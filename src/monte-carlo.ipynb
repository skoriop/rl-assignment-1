```python
import pandas as pd
import random
from collections import defaultdict
from game import Game
import numpy as np
```

```python
actions = [( 1,  0,  0),( 0,  1,  0),(-1,  0,  0),( 0, -1,  0),( 0,  0,  1),(0, 0,  -1),]
```

```python
Q5 = defaultdict(lambda: 0.0)

for action in actions:
    Q5[((1,1,1), action)] = 0.0
```

```python
def generate_epsiode(Q5,eps):
    trajectory = []
    SS = (1,1,1)
    game_instance = Game(50,50)
    while True:
        if np.random.rand() > eps:
            temp1 = {}
            for a in actions:
                temp_pos = tuple(map(lambda i, j: i + j, SS, a))
                # print(temp_pos)
                if game_instance.check(temp_pos) == True:
                  temp1[(SS,(a))] = Q5[(SS,(a))]
            # print(temp1)
            # print(game_instance.position)
            # print(game_instance.turn)
            max_key = max(temp1, key=temp1.get)  
            action = max_key[1]
        else:
            while True:
                action = random.choice(actions)
                temp_pos = tuple(map(lambda i, j: i + j, SS, action))
                if game_instance.check(temp_pos) == True:
                    break
        # print(action)
        new_position = tuple(map(lambda i, j: i + j, SS, action))
        new_position, reward = game_instance.move(new_position)
        trajectory.append((SS, reward, action))
        SS = new_position
        if game_instance.is_terminal(SS):
            break
    return trajectory, len(trajectory) - 1
```

```python
def on_policy_MC(ep, gamma, eps):
    pi = defaultdict(lambda: 1/6)
    Returns = defaultdict(list)
    for _ in range(ep):
        traj, T = generate_epsiode(Q5,eps)
        G = 0
        for t in range(T-1, -1 ,-1):
            s_t, _,a_t = traj[t]
            _,r_t1,_ = traj[t+1]
            G = gamma*G + r_t1
            if not (s_t, a_t) in [(traj[i][0], traj[i][2]) for i in range(0, t)]:
                Returns[(s_t, a_t)].append(G)
                Q5[(s_t, a_t)] = np.average(Returns[(s_t, a_t)])
                temp2 = {}
                for a in actions:
                    temp2[(s_t,(a))] = Q5[(s_t,(a))]
                max_key = max(temp2, key=temp2.get)  
                A_star=max_key[1]
                for a in actions:
                    if a == A_star:   pi[(s_t,a)] = 1 - eps + eps/4
                    else:             pi[(s_t,a)] = eps/4
    return Q5, pi
```

```python
Q_5, pi = on_policy_MC(ep=150, gamma=1, eps=0.1)
```
